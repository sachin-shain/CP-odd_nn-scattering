```python
# choose between 'with contact term' or 'without contact term'


class plot_pot:

    # with contact term
    
    #     folder_plot = 'cp_even_plots/cp_'
    #     folder_c0 = 'c0/c0_'
    
    # without contact term
    
    #     folder_plot = 'cp_even_plots/non_renorm/cp_nr_'
    #     folder_c0 = 'c0/non_renorm/c0_'
    
    #=====================================================================
    
    elab_list = [10,50,100,190]
    dict_key = ['d0_10','d0_50','d0_100','d0_190','d2_10','d2_50','d2_100','d2_190','eps_10','eps_50','eps_100','eps_190']
    
    y_name=[]
    
    def __init__(self,phase_list, folder_plot = 'cp_even_plots/non_renorm/cp_nr_',folder_c0 = 'c0/non_renorm/c0_',lam_div = 1):
        
        
        self.phase_list  = phase_list
        self.folder_plot = folder_plot
        self.folder_c0   = folder_c0
        
        # create filename
        self.c0_filename = ''
        for i in phase_list:
            self.c0_filename += phase_name(*i)+'_'
        self.c0_filename=self.c0_filename
        
        self.c0_name = self.folder_c0+self.c0_filename+'.csv'
        
        # calculate Vtensor0
        self.Vtensor0  = Vtensor_calc(self.phase_list)
        
        self.size = len(phase_list)
        self.lam_div = lam_div
        
        # to determine the number of phase shifts 
        if self.size==1:
            self.y_name = ['d0']
        else:
            self.y_name = ['d0','d2','eps']
        
        # flag_c0  is True if c0 file exists.
        if exists(self.c0_name):
            
            self.df_c0 = pd.read_csv(self.c0_name,index_col=None)
            self.dict_c0 = self.df_c0.to_dict('list')
            self.flag_c0 = True
        
        # if c0(lam) file does not exist, it implies the corresponding contact potential is zero
        else:
            self.dict_c0 ={}
            self.dict_c0['lam'] =[]
            self.dict_c0['c0'] = []
            
            self.flag_c0 = False
            
            lam_list_ref = lambda_values
            
            # setting contact term to zero.
            for lam in lam_list_ref:
                                
                self.dict_c0['lam'].append(lam)
                self.dict_c0['c0'].append(0)
            
            self.df_c0 = pd.DataFrame(self.dict_c0)
#=========================================================================

    def _plot_data(self):

        x = self.dict_c0['lam']
        
        # choice between coupled or uncoupled phases
        def list_gen(n):

            if n==1:

                return [[]]

            elif n==2:

                return [[],[],[]]


        y0 = []

        for elab in self.elab_list:

            print('    {:8}   {:8}   {:8}   {:8}   {:8}   {:8}'.format('elab','lam','c0','d0','d2','e0'))

            q0 = sqrt( elab * m_n * .5)

            Vtensor   = Vtensor_q0_calc(self.phase_list, self.Vtensor0, q0)

            ylist = list_gen(self.size)

            for i in range(len(x)):

                lam = x[i]
                c0 = self.df_c0['c0'][i]

                list_1 = phase_shift_calc(self.phase_list, c0, q0, Vtensor, lam)

                
                print('{:8.5f}   {:8.5f}   {:8.5f}   '.format(elab,lam,c0),end='')


                for j in range(len(list_1)-1):

                    print('{:8.5f}   '.format(list_1[j]),end='')

                    ylist[j].append(list_1[j])
                print()

            y0.append(ylist)
            print()


        y = list_gen(len(phase_list))

        for i in range(len(y)):

            y[i] = []

            for elem in y0:

                y[i].append(elem[i])

        ###############################################

        self.dict_y_list = {}
        self.dict_y_list['lam'] =x
        
        for i in range(len(self.y_name)):

            first =self.y_name[i]+'_'

            for j in range(len(self.elab_list)):
                second = str(self.elab_list[j])

                col_name = first + second
                self.dict_y_list[col_name] = y[i][j]

        self.df_y_list = pd.DataFrame(self.dict_y_list)
        
        # saving the plot data file
        filename = self.folder_plot + self.c0_filename+'.csv'
        self.df_y_list.to_csv(filename,index=False)
        
#========================================================================= 
        
    # plot choice : to see the plots
    # save_choice : to save the plots
    
    def plot(self,plot_choice = True,save_choice = False):
        
        filename =  self.folder_plot + self.c0_filename +'.csv'
        
        if exists(filename):
            
            self.df_y_list = pd.read_csv(filename,index_col=None)
            self.dict_y_list=self.df_y_list.to_dict('list')
        else:
            self._plot_data()
    
        
            
        x = self.dict_y_list['lam']
        
        
        # delta0
        
        legend = []
        for elab in self.elab_list:
            legend.append('Elab = {:} MeV'.format(elab))
            
        
        if plot_choice:
            y = []
            for i in range(0,4):
                y.append(self.dict_y_list[self.dict_key[i]])
                
                
            if save_choice:
                save_fig = [self.folder_plot+'delta0_'+ self.c0_filename+'.pdf',True]
            else:
                save_fig=['',False]
                
            label = ['lambda (fm^-1)','delta0_{:} (degrees)'.format(self.c0_filename)]
            
            display(x,y,label,legend,save=save_fig)
            
        # c0
        
        if self.flag_c0:
            
            if plot_choice:
            
                y = []
                y.append(self.dict_c0['c0'])
                
                if save_choice:
                    save_fig = [self.folder_plot+'c0_'+ self.c0_filename+'.pdf',True]
                else:
                    save_fig=['',False]
                
                
                c0_unit = 'fm^{:d}'.format(self.phase_list[0][1]*2+2)

                label = ['lambda (fm^-1)','c0_{:} ({:})'.format(self.c0_filename,c0_unit)]

                display(x,y,label,legend,save=save_fig)
                
        if self.size==2:
            
            # d2
            
            if plot_choice:
                y = []
                for i in range(4,8):
                    y.append(self.dict_y_list[self.dict_key[i]])
                    
                if save_choice:
                    save_fig = [self.folder_plot+'delta2_'+ self.c0_filename+'.pdf',True]
                else:
                    save_fig=['',False]


                label = ['lambda (fm^-1)','delta2_{:} (degrees)'.format(self.c0_filename)]

                display(x,y,label,legend,save=save_fig)
            
            # eps
            
            
            if plot_choice:
                y = []
                for i in range(8,12):
                    y.append(self.dict_y_list[self.dict_key[i]])
                
                if save_choice:
                    save_fig = [self.folder_plot+'eps_'+ self.c0_filename+'.pdf',True]
                else:
                    save_fig=['',False]


                label = ['lambda (fm^-1)','eps_{:} (degrees)'.format(self.c0_filename)]

                display(x,y,label,legend,save=save_fig)
                
            
            
        
        
```

```python
phase_list = [[0,0,0]]

p_wave = [phase_name(*phase) for phase in phase_list]
channel = ''
for phase in p_wave:
    channel+= phase + " "
print('CP-even plots for channels : ',channel)

print('\n\nWithout contact term: \n\n')

folder_plot = 'cp_even_plots/non_renorm/cp_nr_'
folder_c0 = 'c0/non_renorm/c0_'

s1 = plot_pot(phase_list,folder_plot,folder_c0)
s1.plot(True,True)

print('\n\nWith contact term: \n\n')

folder_plot = 'cp_even_plots/cp_'
folder_c0 = 'c0/c0_'

s1 = plot_pot(phase_list,folder_plot,folder_c0)
s1.plot(True,True)
```
