```python
class plot_comp:
    
    folder_plot = 'comparison_plots/comp_'
    folder_true= 'comparison_plots/data/true_'
    folder_c0 = 'c0/c0_'
    
    dict_key = ['elab','d0','d0_true','d2','d2_true','e0','e0_true']
    
    data_y_dict_cols=['lam','elab','d0','d0_true','d2','d2_true','e0','e0_true']
    
    y_name=[]
    
    def __init__(self,phase_list):
        
        self.phase_list = phase_list
        
        # generate filename
        self.c0_filename = ''
        for i in phase_list:
            self.c0_filename += phase_name(*i)+'_'
        
        self.c0_name = self.folder_c0+self.c0_filename+'.csv'
        
        # calculate Vtensor0
        self.Vtensor0  = Vtensor_calc(self.phase_list)
        
        self.size = len(phase_list)
        
        self.data_y_dict ={'lam':[],'elab':[],'d0':[],'d0_true':[]}
        
        # choice between coupled and uncoupled phases
        if self.size==1:
            self.y_name = ['d0']
        else:
            self.y_name = ['d0','d2','eps']
            for i in range(4,len(self.data_y_dict_cols)):
                self.data_y_dict[self.data_y_dict_cols[i]]=[]
                

        # flag_c0 is True if c0 file exists.
        # otherwise, assume c0 = 0.
        if exists(self.c0_name):

            self.df_c0 = pd.read_csv(self.c0_name,index_col=None)
            self.dict_c0 = self.df_c0.to_dict('list')
            self.flag_c0 = True           
            
            
        else:
            self.dict_c0 ={}
            self.dict_c0['lam'] =[]
            self.dict_c0['c0'] = []
            
            self.flag_c0 = False
            
            lam_list_ref = deepcopy(lambda_values)
            
            for lam in lam_list_ref:
                                
                self.dict_c0['lam'].append(lam)
                self.dict_c0['c0'].append(0)
            
            self.df_c0 = pd.DataFrame(self.dict_c0)
            
        self.true_data_name = self.folder_true +self.c0_filename+'.csv'
        
        if exists(self.true_data_name):
            
            print('\nExperimental data file found.')

            self.df_true = pd.read_csv(self.true_data_name,index_col=None)
            self.dict_true = self.df_true.to_dict('list')
        else:
            print('\nExperimental data file not found.')
            
#=========================================================================

    def _plot_data(self):

        x = self.dict_c0['lam']
        
        # choice between coupled and uncoupled
        def list_gen(n):

            if n==1:

                return [[]]

            elif n==2:

                return [[],[],[]]


        y0 = {}
        
        print('    {:8}   {:8}   {:8}   {:8}   {:8}   {:8}   {:8}   {:8}'.format('lam','elab','d0','d0_true','d2','d2_true','e0','e0_true'))
        
        
        #==================================================
        #                    for loop
        #==================================================
        
        # we choose lam = 52 fm^-1 for the plots
        
        for i in self.dict_true:
            print(i)
        input()
        for elab in list(self.dict_true['elab']):
            
            
            elab_index = self.dict_true['elab'].index(elab)
            q0 = sqrt( elab * m_n * .5)

            Vtensor   = Vtensor_q0_calc(self.phase_list, self.Vtensor0, q0)
    
            lam_index = self.dict_c0['lam'].index(52)
            lam = 52.0
            
            c0 = self.df_c0['c0'][lam_index]

            list_1 = phase_shift_calc(self.phase_list, c0, q0, Vtensor, lam)

            print('{:8.5f}   {:8.5f}   '.format(lam,elab),end='')
            
            ylist = [lam,elab]

            for j in range(len(list_1)-1):
                
                true_val =self.dict_true[self.dict_key[2+j*2]][elab_index]

                print('{:8.5f}   {:8.5f}   '.format(list_1[j],true_val),end='')
                ylist.append(list_1[j])
                ylist.append(true_val)
                
            print()
            
            for i in range(len(ylist)):
                self.data_y_dict[self.data_y_dict_cols[i]].append(ylist[i])
        


#         ###############################################

        
        self.data_y_df = pd.DataFrame(self.data_y_dict)

        filename = self.folder_plot + self.c0_filename + '.csv'
        self.data_y_df.to_csv(filename,index=False)
        
        
#========================================================================= 

    def plot(self):
        
        data_file_name =self.folder_plot + self.c0_filename + '.csv'
        
        if exists(data_file_name):
            
            print('C0 data file found.\n')
        else:
            
            self._plot_data()
            
        
        self.data_y_df = pd.read_csv(data_file_name,index_col=None)
        self.data_y_dict = self.data_y_df.to_dict('list')
        
        x = self.data_y_dict['elab']
        legend = ['LO','True']
        
        # delta 0
        
        y = []
        y.append(self.data_y_dict['d0'])
        y.append(self.data_y_dict['d0_true'])
        
        
        save_fig = [self.folder_plot+'d0_'+ self.c0_filename+'.pdf',True]
        
        label = ['elab (MeV)','{:} (degrees)'.format(self.c0_filename + 'd0')]
        
        display(x,y,label,legend,save=save_fig)
        
        if len(self.data_y_dict)==8:
        
            # delta 2

            y = []
            y.append(self.data_y_dict['d2'])
            y.append(self.data_y_dict['d2_true'])


            save_fig = [self.folder_plot+ 'd2_'+self.c0_filename+'.pdf',True]

            label = ['elab (MeV)','{:} (degrees)'.format(self.c0_filename + 'd2')]

            display(x,y,label,legend,save=save_fig)
            
            # elsilon

            y = []
            y.append(self.data_y_dict['e0'])
            y.append(self.data_y_dict['e0_true'])


            save_fig = [self.folder_plot+ 'e0_'+self.c0_filename+'.pdf',True]

            label = ['elab (MeV)','{:} (degrees)'.format(self.c0_filename + 'e0')]

            display(x,y,label,legend,save=save_fig)
            
        
                
                
        
        
        
        
```
