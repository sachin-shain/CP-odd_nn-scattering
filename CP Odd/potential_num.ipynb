### $g_n^f(p',p) = \int_{-1}^1 dx P_n (x) F_f(q(x,p',p)) $

where,
### $F_f[q(x,p',p)] = \frac{q^{2-f}}{q^2 + m_\pi^2}, \hspace{1cm} q(x,p',p) = \sqrt{p'^2 + p^2 - 2 p' p x}$

```python
# g function for calculating potential

# f =0, 2
# n = 0,...,5

def g_fun(f,n,i1,i2):

    p1 = pgrid[i1]
    p2 = pgrid[i2]

    if f==0 and n==0:
        return  (
                    2 - ( mpi**2
                          *   atanh( (2 * p1 * p2)/( mpi**2 + p1**2 + p2**2 )  )
                        )/ (p1 * p2)
                )
    elif f==0 and n==1:
        return  (
                    (
                        mpi**2 * (  4 * p1 * p2 
                                    - ( 2 * (mpi**2 + p1**2 + p2**2) 
                                        *  atanh( (2 * p1 * p2)/( mpi**2 + p1**2 + p2**2 )  )
                                      )
                                 )
                    )/(4 * p1**2 * p2**2)
                )
    elif f==0 and n==2:
        return  (
                    (
                        mpi**2 
                        * (
                            12 * p1 * p2 * (mpi**2 + p1**2 + p2**2) 
                            - 2 * (3 * (mpi**2 + p1**2)**2  + 2 * (3* mpi**2 + p1**2) * p2**2 + 3 * p2**4   )
                                * atanh( (2 * p1 * p2)/( mpi**2 + p1**2 + p2**2 )  )
                          )
                    )/(16 * p1**3 * p2**3)
                )
    
    elif f==0 and n==3:
        return  (
                    ( mpi**2 / ( 48 * p1**4 * p2**4 ) ) * 
                        (
                            30 * p1 * ( mpi**2 + p1**2 )**2 * p2 
                            + 4 * p1 * ( 15 * mpi**2 + 7 * p1**2 ) * p2**3
                            + 30 * p1 * p2**5
                            - 3 * ( mpi**2 + p1**2 + p2**2 )
                                * ( 5 * ( mpi**2 + p1**2 )**2 + 2 * ( 5 * mpi**2 - p1**2 ) * p2**2 + 5 * p2**4 ) 
                                * atanh( 2 * p1 * p2 / ( mpi**2 + p1**2 + p2**2 ) )
                        )
                )

    elif f==0 and n==4:
        return  (
                    ( mpi**2 / ( 768 * p1**5 * p2**5 ) ) 
                    * (
                        20 * p1 * p2 * ( mpi**2 + p1**2 + p2**2 ) * ( 21 * ( mpi**2 + p1**2 )**2 + 2 * ( 21 * mpi**2 - p1**2 ) * p2**2 + 21 * p2**4 )
                        - 6 * ( 
                                35 * ( mpi**2 + p1**2 )**4 
                                + 20 * ( mpi**2 + p1**2 )**2 * ( 7 * mpi**2 + p1**2 ) * p2**2 
                                + 6 * ( 35 * mpi**4 + 30 * mpi**2 * p1**2 + 3 * p1**4 ) * p2**4
                                + 20 * ( 7 * mpi**2 + p1**2 ) * p2**6
                                + 35 * p2**8
                              )
                        * atanh( 2 * p1 * p2 / ( mpi**2 + p1**2 + p2**2 ) )
                      )
                )
                
    elif f==2 and n==0:
        return  (
                    log(  1 
                          + (4 * p1 * p2)/( mpi**2 + (p1 - p2)**2 )
                       )/( 2* p1 * p2)
                )          
    elif f==2 and n==1:
        return  (
                    (
                        - 2 * p1 * p2
                        + (
                            (mpi**2 + p1**2 + p2**2 )
                            *  atanh( (2 * p1 * p2)/( mpi**2 + p1**2 + p2**2 )  )
                          )
                    )/(2 * p1**2 * p2**2)
                )
    elif f==2 and n==2:
        return  (
                    (
                        3
                        * (mpi**2 + p1**2 + p2**2)
                        * (
                            -2 * p1 * p2 + (mpi**2 + p1**2 + p2**2) *  atanh( (2 * p1 * p2)/( mpi**2 + p1**2 + p2**2 )  )
                          )
                        - 2 * p1**2 * p2**2 
                          * log( 1 + (4 * p1 * p2)/(mpi**2  + (p1-p2)**2) )
                    )/(8 * p1**3 * p2**3)
                )
    elif f==2 and n==3:
        return  (
                    -(
                        30 * p1 * (mpi**2 + p1**2)**2 * p2
                        + 4 * p1 * (15 * mpi**2 + 7 * p1**2) * p2**3
                        + 30 * p1 * p2**5
                        - 3 * (mpi**2 + p1**2 + p2**2) 
                          * (
                              5 * (mpi**2 + p1**2)**2 + 2 * (5 * mpi**2 - p1**2) * p2**2 + 5 * p2**4
                            )
                          * atanh( (2 * p1 * p2)/( mpi**2 + p1**2 + p2**2 )  )
                     )/(48 * p1**4 * p2**4)
                )
    elif f==2 and n==4:
        return  (
                    (768 * p1**5 * p2**5)**-1 
                    * (
                            -20 * p1 * p2 * ( mpi**2 + p1**2 + p2**2 ) 
                              * ( 21 * ( mpi**2 + p1**2 )**2 + 2 * ( 21 * mpi**2 - p1**2 ) * p2**2 + 21 * p2**4 ) 
                            + 6 * ( 
                                    35 * ( mpi**2 + p1**2 )**4 
                                    + 20 * ( mpi**2 + p1**2 )**2 * ( 7 * mpi**2 + p1**2 ) * p2**2
                                    + 6 * ( 35 * mpi**4 + 30 * mpi**2 * p1**2 + 3 * p1**4 ) * p2**4
                                    + 20 * ( 7 * mpi**2 + p1**2 ) * p2**6
                                    + 35 * p2**8
                                    )
                                * atanh( (2 * p1 * p2)/( mpi**2 + p1**2 + p2**2 ) )
                      )
                )
    elif f==2 and n==5:
        return  (
                    4 * p1 * p2 
                        * ( 
                            -945 * ( mpi**2 + p1**2 )**4
                            -420 * ( mpi**2 + p1**2 )**2 * ( 9 * mpi**2 + 2 * p1**2 ) * p2**2
                            -2 * ( 2835 * mpi**4 + 2730 * mpi**2 * p1**2 + 407 * p1**4 ) * p2**4
                            -420 * ( 9 * mpi**2 + 2 * p1**2 ) * p2**6
                            -945 * p2**8 
                          )
                    + 30 * ( mpi**2 + p1**2 + p2**2 )
                       *  (
                             63 * ( mpi**2 + p1**2 )**4
                             + 28 * ( 9 * mpi**2 - p1**2 ) * ( mpi**2 + p1**2 )**2 * p2**2
                             + 2 * ( 189 * mpi**4 + 98 * mpi**2 * p1**2 + 29 * p1**4 ) * p2**4
                             + 28 * ( 9 * mpi**2 - p1**2 ) * p2**6
                             + 63 * p2**8
                          ) * atanh( (2 * p1 * p2)/( mpi**2 + p1**2 + p2**2 ) )
                ) * (7680 * p1**6 * p2**6)**-1
    else:
        print('Have to type the g function corresponds to f={:d}, n={:d}'.format(f,n))
        input('\nEnter any key to continue...')
        return 0
```

### $V(p',p) = \langle p; (l',s')j'm_j';t' m_t'|V_{1\pi}|p;(ls)jm_j;tm_t \rangle $

```python
# CP-even potential for a given s1,l1,j1---s2,l2,j2 

def pot_even_one_pion(s1,l1,j1,s2,l2,j2,i1,i2):
    
    A0  = -(gA/(2 * fpi))**2 * (2*pi)**(-3)

    p1  = pgrid[i1]
    p2  = pgrid[i2]

#     p1 =200
#     p2=200



    # j = 0     : 1S0-3P0
    
    '''
        non-zero:   1S0 - 1S0
        
                    3P0 - 3P0
    '''


    if [s1,l1,j1]==[0,0,0] and [s2,l2,j2]==[0,0,0]:         # 1S0 - 1S0 
        return   -2  * A0 * pi * g_fun(0,0,i1,i2)

    elif [s1,l1,j1]==[0,0,0] and [s2,l2,j2]==[1,1,0]:       # 1S0 - 3P0
        return   0

    elif [s1,l1,j1]==[1,1,0] and [s2,l2,j2]==[0,0,0]:       # 3P0 - 1S0
        return   0

    elif [s1,l1,j1]==[1,1,0] and [s2,l2,j2]==[1,1,0]:       # 3P0 - 3P0
        return   (
                    (2/9) * A0 * pi
                    * (
                            3 * g_fun(0,1,i1,i2)
                            + 4 * (
                                    5 * p1 * p2 * g_fun(2,0,i1,i2)
                                    -3 * (p1**2 + p2**2) * g_fun(2,1,i1,i2)
                                    + p1 * p2 * g_fun(2,2, i1,i2) 
                                )
                        )
                )

    
#     j = 1     : 3S1-3D1-3P1-1P1
    
    '''
        non-zero:   3S1 - 3S1
                    3S1 - 3D1
                    3D1 - 3S1
                    3D1 - 3D1
                    
                    3P1 - 3P1
                    
                    1P1 - 1P1
    '''
    
    if [s1,l1,j1]==[1,0,1] and [s2,l2,j2]==[1,0,1]:       # 3S1 - 3S1
        return   -2 * A0 * pi * g_fun(0,0,i1,i2)

    elif [s1,l1,j1]==[1,0,1] and [s2,l2,j2]==[1,2,1]:       # 3S1 - 3D1
        return   (
                    - 4 * sqrt(2) * A0 * pi 
                        * (
                            p2**2 * g_fun(2,0,i1,i2)
                            - 2 * p1 * p2 * g_fun(2,1,i1,i2)
                            + p1**2 * g_fun(2,2,i1,i2)
                          )
                )

    elif [s1,l1,j1]==[1,0,1] and [s2,l2,j2]==[1,1,1]:       # 3S1 - 3P1
        return  0
    
    elif [s1,l1,j1]==[1,0,1] and [s2,l2,j2]==[0,1,1]:       # 3S1 - 1P1
        return  0

    elif [s1,l1,j1]==[1,2,1] and [s2,l2,j2]==[1,0,1]:       # 3D1 - 3S1
        return   (
                    - 4 * sqrt(2) * A0 * pi 
                        * (
                            p1**2 * g_fun(2,0,i1,i2)
                            - 2 * p1 * p2 * g_fun(2,1,i1,i2)
                            + p2**2 * g_fun(2,2,i1,i2)
                          )
                )
    
    elif [s1,l1,j1]==[1,2,1] and [s2,l2,j2]==[1,2,1]:       # 3D1 - 3D1
        return   (
                    -(2/5) * A0 * pi
                    * (
                        5 * g_fun(0,2,i1,i2)
                        + 14 * p1 * p2 * g_fun(2,1,i1,i2)
                        - 10 * (p1**2 + p2**2) * g_fun(2,2,i1,i2)
                        + 6 * p1 * p2 * g_fun(2,3,i1,i2)  
                      )
                )

    elif [s1,l1,j1]==[1,2,1] and [s2,l2,j2]==[1,1,1]:       # 3D1 - 3P1
        return  0

    elif [s1,l1,j1]==[1,2,1] and [s2,l2,j2]==[0,1,1]:       # 3D1 - 1P1
        return  0

    elif [s1,l1,j1]==[1,1,1] and [s2,l2,j2]==[1,0,1]:       # 3P1 - 3S1
        return  0
    
    elif [s1,l1,j1]==[1,1,1] and [s2,l2,j2]==[1,2,1]:       # 3P1 - 3D1
        return  0

    elif [s1,l1,j1]==[1,1,1] and [s2,l2,j2]==[1,1,1]:       # 3P1 - 3P1
        return   (
                    (2/9) * A0 * pi
                    * (
                            3 * g_fun(0,1,i1,i2)
                            -10 * p1 * p2 * g_fun(2,0,i1,i2)
                            + 6 * (p1**2 + p2**2) * g_fun(2,1, i1,i2)
                            - 2 * p1 * p2 * g_fun(2,2,i1,i2)
                      )
                )

    elif [s1,l1,j1]==[1,1,1] and [s2,l2,j2]==[0,1,1]:       # 3P1 - 1P1
        return  0

    elif [s1,l1,j1]==[0,1,1] and [s2,l2,j2]==[1,0,1]:       # 1P1 - 3S1
        return  0
    
    elif [s1,l1,j1]==[0,1,1] and [s2,l2,j2]==[1,2,1]:       # 1P1 - 3D1
        return  0

    elif [s1,l1,j1]==[0,1,1] and [s2,l2,j2]==[1,1,1]:       # 1P1 - 3P1
        return  0

    elif [s1,l1,j1]==[0,1,1] and [s2,l2,j2]==[0,1,1]:       # 1P1 - 1P1
        return   6 * A0 * pi * g_fun(0,1,i1,i2)


    # j = 2     : 1D2-3P2-3D2-3F2
    
    '''
        non-zero:   1D2 - 1D2
        
                    3D2 - 3D2
        
                    3P2 - 3P2
                    3P2 - 3F2
                    3F2 - 3P2                    
                    3F2 - 3F2
    '''


    if [s1,l1,j1]==[0,2,2] and [s2,l2,j2]==[0,2,2]:       # 1D2 - 1D2
        return  -2 * A0 * pi * g_fun(0,2,i1,i2)

    elif [s1,l1,j1]==[0,2,2] and [s2,l2,j2]==[1,1,2]:       # 1D2 - 3P2
        return 0

    elif [s1,l1,j1]==[0,2,2] and [s2,l2,j2]==[1,2,2]:       # 1D2 - 3D2
        return 0

    elif [s1,l1,j1]==[0,2,2] and [s2,l2,j2]==[1,3,2]:       # 1D2 - 3F2
        return 0

    elif [s1,l1,j1]==[1,1,2] and [s2,l2,j2]==[0,2,2]:       # 3P2 - 1D2
        return 0

    elif [s1,l1,j1]==[1,1,2] and [s2,l2,j2]==[1,1,2]:       # 3P2 - 3P2
        return  (
                    2/45 * pi * A0
                    *   (
                            15 * g_fun(0,1,i1,i2) + 10 * p1 * p2 * g_fun(2,0,i1,i2) 
                            - 6 * p1**2 * g_fun(2,1,i1,i2) - 6 * p2**2 * g_fun(2,1,i1,i2)
                            + 2 * p1 * p2 * g_fun(2,2,i1,i2)
                        )
                )

    elif [s1,l1,j1]==[1,1,2] and [s2,l2,j2]==[1,2,2]:       # 3P2 - 3D2
        return 0

    elif [s1,l1,j1]==[1,1,2] and [s2,l2,j2]==[1,3,2]:       # 3P2 - 3F2
        return  (
                    4/5 * sqrt(6) * pi * A0
                    *   (
                            p2**2  * g_fun(2,1,i1,i2) - 2 * p1 * p2 * g_fun(2,2,i1,i2) 
                            + p1**2 * g_fun(2,3,i1,i2)
                        )
                )

    elif [s1,l1,j1]==[1,2,2] and [s2,l2,j2]==[0,2,2]:       # 3D2 - 1D2
        return 0

    elif [s1,l1,j1]==[1,2,2] and [s2,l2,j2]==[1,1,2]:       # 3D2 - 3P2
        return 0

    elif [s1,l1,j1]==[1,2,2] and [s2,l2,j2]==[1,2,2]:       # 3D2 - 3D2
        return  (
                    -2/5 * pi * A0
                    *   (
                            5  * g_fun(0,2,i1,i2) - 14 * p1 * p2 * g_fun(2,1,i1,i2) 
                            + 10 * p1**2 * g_fun(2,2,i1,i2) + 10 * p2**2 * g_fun(2,2,i1,i2)
                            - 6 * p1 * p2 * g_fun(2,3,i1,i2)
                        )
                )

    elif [s1,l1,j1]==[1,2,2] and [s2,l2,j2]==[1,3,2]:       # 3D2 - 3F2
        return 0
    
    elif [s1,l1,j1]==[1,3,2] and [s2,l2,j2]==[0,2,2]:       # 3F2 - 1D2
        return 0

    elif [s1,l1,j1]==[1,3,2] and [s2,l2,j2]==[1,1,2]:       # 3F2 - 3P2
        return  (
                    4/5 * sqrt(6) * pi * A0
                    *   (
                            p1**2  * g_fun(2,1,i1,i2) - 2 * p1 * p2 * g_fun(2,2,i1,i2) 
                            + p2**2 * g_fun(2,3,i1,i2)
                        )
                )

    elif [s1,l1,j1]==[1,3,2] and [s2,l2,j2]==[1,2,2]:       # 3F2 - 3D2
        return 0

    elif [s1,l1,j1]==[1,3,2] and [s2,l2,j2]==[1,3,2]:       # 3F2 - 3F2
        return  (
                    2/105 * pi * A0
                    *   (
                            35  * g_fun(0,3,i1,i2) + 72 * p1 * p2 * g_fun(2,2,i1,i2) 
                            - 56 * p1**2 * g_fun(2,3,i1,i2) - 56 * p2**2 * g_fun(2,3,i1,i2)
                            + 40 * p1 * p2 * g_fun(2,4,i1,i2)
                        )
                )


    else:
        print('Have to type CP-even the potential corresponds to {}-{} .'.format(phase_name(s1,l1,j1),phase_name(s2,l2,j2)))
        input('\nEnter any key to continue...')
        return 'exit'
```

### $V(p',p) = \langle p; (l',s')j'm_j';t' m_t'|V_{contact}(c_0)|p;(ls)jm_j;tm_t \rangle $

```python

def pot_even_contact(s1,l1,j1,s2,l2,j2,i1,i2,c0=0):
    
    if [s1,l1,j1] == [0,0,0] and [s2,l2,j2] == [0,0,0]:     # 1S0-1S0
        return   ( c0/( (2 * pi)**3 )) * hc**-2

    elif [s1,l1,j1] == [1,0,1] and [s2,l2,j2] == [1,0,1]:   # 3S1-3S1
        return  ( c0/(  (2 * pi)**3 )) * hc**-2

    elif [s1,l1,j1] == [1,1,0] and [s2,l2,j2] == [1,1,0]:   # 3P0-3P0
        return   (
                    ( c0/( 4 *(2 * pi)**3 ))
                    * pgrid[i1]
                    * pgrid[i2]
                ) * hc**-4

    elif [s1,l1,j1] == [1,1,2] and [s2,l2,j2] == [1,1,2]:   # 3P2-3P2
        return   (
                    ( c0/( 4 *(2 * pi)**3 ))
                    * pgrid[i1]
                    * pgrid[i2]
                ) * hc**-4

    elif [s1,l1,j1] == [1,2,2] and [s2,l2,j2] == [1,2,2]:   # 3D2-3D2
        return    (
                    c0 * (2 * pi)**-3
                    * pgrid[i1]**2
                    * pgrid[i2]**2
                ) * hc**-6

    else:
        return 0
```

## $V^{even}_{(N_p)\times(N_p)}$

```python

class pot_even_Dictionary:
    
    def __init__(self):
        self._cache = Vtensor_even_ref

    def __call__(self, *args):
        
        [s1,l1,j1,s2,l2,j2,i1,i2] = [S(i)  for i in args]
        
        if i1==Np or i2==Np:
            return pot_even_one_pion(*[S(i)  for i in args])
        
        res = self._cache.get(args, None)  # return None if args keyword does not exist
        if res is None:
            res =  pot_even_one_pion(*[S(i)  for i in args])
        
            self._cache[args] = res
        
        return res
    
    def read(self):
        return self._cache
        
    def delete(self):
        self._cache = {}
        
    def show(self,choice = False):
        print('length =', len(self._cache))
        
        if choice:
            print_(self._cache,dataframe=True)
        return len(self._cache)
    
pot_even_dict = pot_even_Dictionary()
```

## $$V_{cp-odd} = -i\frac{g_0 g_A}{2 F_\pi} \frac{(\sigma_1-\sigma_2)\cdot q}{q^2 + m_\pi^2}\tau_1\cdot \tau_2$$

### $V(p',p) = \langle p; (l',s')j'm_j';t' m_t'|V_{odd,1\pi}|p;(ls)jm_j;tm_t \rangle $

```python
# CP-odd potential for a given s1,l1,j1---s2,l2,j2 
def pot_cp_odd_one_pion(s1,l1,j1,s2,l2,j2,i1,i2):

    
    A1  = -1j * (gA * g0)/(2 * fpi) * (2*pi)**(-3)

    p1  = pgrid[i1]
    p2  = pgrid[i2]

    # j = 0     : 1S0-1P0
    
    '''
        non-zero:   1S0 - 3P0
                    3P0 - 1S0
    '''

    if [s1,l1,j1]==[0,0,0] and [s2,l2,j2]==[0,0,0]:         # 1S0 - 1S0 
        return   0

    elif [s1,l1,j1]==[0,0,0] and [s2,l2,j2]==[1,1,0]:       # 1S0 - 3P0
        return    ( 
                    A1 * 4 * pi 
                    * ( p2 * g_fun(2,0,i1,i2) 
                        - p1 * g_fun(2,1,i1,i2) 
                      )
                  )

    elif [s1,l1,j1]==[1,1,0] and [s2,l2,j2]==[0,0,0]:       # 3P0 - 1S0
        return    -( 
                    A1 * 4 * pi 
                    * ( p1 * g_fun(2,0,i1,i2) 
                        - p2 * g_fun(2,1,i1,i2) 
                      )
                  )

    elif [s1,l1,j1]==[1,1,0] and [s2,l2,j2]==[1,1,0]:       # 3P0 - 3P0
        return 0


    # j = 1     : 3S1-3D1-3P1-1P1
    
    
    '''
        non-zero:   3S1 - 1P1
                    1P1 - 3S1
                    3D1 - 1P1
                    1P1 - 3D1
    '''
    
    if [s1,l1,j1]==[1,0,1] and [s2,l2,j2]==[1,0,1]:         # 3S1 - 3S1
        return  0

    elif [s1,l1,j1]==[1,0,1] and [s2,l2,j2]==[1,2,1]:       # 3S1 - 3D1
        return  0

    elif [s1,l1,j1]==[1,0,1] and [s2,l2,j2]==[1,1,1]:       # 3S1 - 3P1
        return  0
    
    elif [s1,l1,j1]==[1,0,1] and [s2,l2,j2]==[0,1,1]:       # 3S1 - 1P1
        return  ( 
                    A1 * 4 * sqrt(3) * pi 
                    * ( p2 * g_fun(2,0,i1,i2) 
                        - p1 * g_fun(2,1,i1,i2) 
                      )
                )          

    elif [s1,l1,j1]==[1,2,1] and [s2,l2,j2]==[1,0,1]:       # 3D1 - 3S1
        return  0
    
    elif [s1,l1,j1]==[1,2,1] and [s2,l2,j2]==[1,2,1]:       # 3D1 - 3D1
        return  0

    elif [s1,l1,j1]==[1,2,1] and [s2,l2,j2]==[1,1,1]:       # 3D1 - 3P1
        return  0

    elif [s1,l1,j1]==[1,2,1] and [s2,l2,j2]==[0,1,1]:       # 3D1 - 1P1
        return  (
                    A1 * 4. * pi * sqrt(6.)
                    * (
                        p1 * g_fun(2,1,i1,i2)
                        - p2 * g_fun(2,2,i1,i2)
                      )
                 )

    elif [s1,l1,j1]==[1,1,1] and [s2,l2,j2]==[1,0,1]:       # 3P1 - 3S1
        return  0
    
    elif [s1,l1,j1]==[1,1,1] and [s2,l2,j2]==[1,2,1]:       # 3P1 - 3D1
        return  0

    elif [s1,l1,j1]==[1,1,1] and [s2,l2,j2]==[1,1,1]:       # 3P1 - 3P1
        return  0

    elif [s1,l1,j1]==[1,1,1] and [s2,l2,j2]==[0,1,1]:       # 3P1 - 1P1
        return  0

    elif [s1,l1,j1]==[0,1,1] and [s2,l2,j2]==[1,0,1]:       # 1P1 - 3S1
        return  -( 
                    A1 * 4 * sqrt(3) * pi 
                    * ( p1 * g_fun(2,0,i1,i2) 
                        - p2 * g_fun(2,1,i1,i2) 
                      )
                )
    
    elif [s1,l1,j1]==[0,1,1] and [s2,l2,j2]==[1,2,1]:       # 1P1 - 3D1
        return  -(
                    A1 * 4. * pi * sqrt(6.)
                    * (
                        p2 * g_fun(2,1,i1,i2)
                        - p1 * g_fun(2,2,i1,i2)
                      )
                )

    elif [s1,l1,j1]==[0,1,1] and [s2,l2,j2]==[1,1,1]:       # 1P1 - 3P1
        return  0

    elif [s1,l1,j1]==[0,1,1] and [s2,l2,j2]==[0,1,1]:       # 1P1 - 1S0
        return  0

    # j = 2     : 1D2-3P2-3D2-3F2
    
    
    '''
        non-zero:   1D2 - 3P2
                    3P2 - 1D2
                    
                    1D2 - 3F2
                    3F2 - 1D2
    '''


    if [s1,l1,j1]==[0,2,2] and [s2,l2,j2]==[0,2,2]:       # 1D2 - 1D2
        return  0

    elif [s1,l1,j1]==[0,2,2] and [s2,l2,j2]==[1,1,2]:       # 1D2 - 3P2
        return  (
                    4 * sqrt(2/5) * pi * A1
                    *   (
                            p1  * g_fun(2,1,i1,i2) - p2 * g_fun(2,2,i1,i2) 
                        )
                )

    elif [s1,l1,j1]==[0,2,2] and [s2,l2,j2]==[1,2,2]:       # 1D2 - 3D2
        return  0

    elif [s1,l1,j1]==[0,2,2] and [s2,l2,j2]==[1,3,2]:       # 1D2 - 3F2
        return  (
                    4 * sqrt(3/5) * pi * A1
                    *   (
                            p2  * g_fun(2,2,i1,i2) - p1 * g_fun(2,3,i1,i2) 
                        )
                )

    elif [s1,l1,j1]==[1,1,2] and [s2,l2,j2]==[0,2,2]:       # 3P2 - 1D2
        return  (
                    -4 * sqrt(2/5) * pi * A1
                    *   (
                            p2  * g_fun(2,1,i1,i2) - p1 * g_fun(2,2,i1,i2) 
                        )
                )

    elif [s1,l1,j1]==[1,1,2] and [s2,l2,j2]==[1,1,2]:       # 3P2 - 3P2
        return 0

    elif [s1,l1,j1]==[1,1,2] and [s2,l2,j2]==[1,2,2]:       # 3P2 - 3D2
        return 0

    elif [s1,l1,j1]==[1,1,2] and [s2,l2,j2]==[1,3,2]:       # 3P2 - 3F2
        return 0 

    elif [s1,l1,j1]==[1,2,2] and [s2,l2,j2]==[0,2,2]:       # 3D2 - 1D2
        return 0

    elif [s1,l1,j1]==[1,2,2] and [s2,l2,j2]==[1,1,2]:       # 3D2 - 3P2
        return 0

    elif [s1,l1,j1]==[1,2,2] and [s2,l2,j2]==[1,2,2]:       # 3D2 - 3D2
        return 0

    elif [s1,l1,j1]==[1,2,2] and [s2,l2,j2]==[1,3,2]:       # 3D2 - 3F2
        return 0
    
    elif [s1,l1,j1]==[1,3,2] and [s2,l2,j2]==[0,2,2]:       # 3F2 - 1D2
        return  (
                    -4 * sqrt(3/5) * pi * A1
                    *   (
                            p1  * g_fun(2,2,i1,i2) - p2 * g_fun(2,3,i1,i2) 
                        )
                )

    elif [s1,l1,j1]==[1,3,2] and [s2,l2,j2]==[1,1,2]:       # 3P2 - 3P2
        return 0

    elif [s1,l1,j1]==[1,3,2] and [s2,l2,j2]==[1,2,2]:       # 3F2 - 3D2
        return 0

    elif [s1,l1,j1]==[1,3,2] and [s2,l2,j2]==[1,3,2]:       # 3F2 - 3F2
        return 0

    else:
        print('Have to type CP-odd the potential corresponds to {}-{} .'.format(phase_name(s1,l1,j1),phase_name(s2,l2,j2)))
        input('\nEnter any key to continue...')
        return 'exit'
```

### $V(p',p) = \langle p; (l',s')j'm_j';t' m_t'|V_{cp-odd,contact}(c_0)|p;(ls)jm_j;tm_t \rangle $

```python
def pot_cp_odd_contact(s1,l1,j1,s2,l2,j2,i1,i2):


    c0 = 1./hc**3 
    B1  = 1j * c0 * (2*pi)**(-3)

    p1  = pgrid[i1]
    p2  = pgrid[i2]

    # j = 0     : 1S0-1P0

    if [s1,l1,j1]==[0,0,0] and [s2,l2,j2]==[1,1,0]:         # 1S0 - 3P0 
        return 8 * p2 * pi * B1

    elif [s1,l1,j1]==[1,1,0] and [s2,l2,j2]==[0,0,0]:       # 3P0 - 1S0
        return -8 * p1 * pi * B1
    
    else:
        return 0
```

## $V^{cp-odd}_{(N_p)\times(N_p)}$

```python
class pot_cp_odd_Dictionary:
    
    def __init__(self):
        self._cache = Vtensor_odd_ref

    def __call__(self, *args):
        
        [s1,l1,j1,s2,l2,j2,i1,i2] = [S(i)  for i in args]
        
        if i1==Np or i2==Np:
            return pot_cp_odd_one_pion(*[S(i)  for i in args])
        
        res = self._cache.get(args, None)  # return None if args keyword does not exist
        if res is None:
            res =  pot_cp_odd_one_pion(*[S(i)  for i in args])
        
            self._cache[args] = res
        
        return res
    
    def read(self):
        return self._cache
        
    def delete(self):
        self._cache = {}
        
    def show(self,choice = False):
        print('length =', len(self._cache))
        
        if choice:
            print_(self._cache,dataframe=True)
        return len(self._cache)
    
pot_cp_odd_dict = pot_cp_odd_Dictionary()

def write_file(p_even, p_odd):
    
    Vfile.write(p_even.read(),p_odd.read())
```

## Potential for various index:

| index      | potetential |
| ---| ----------- |
| ...........     |  .................................................. |
| 0       | $V_{even} $                                  |
| 1       | $\epsilon V_{odd} $                                   |
| 2       | $V_{even} + \epsilon V_{odd} $               |
| 3       | $V_{even} + \epsilon V_{contact}$            |
| 4       | $V_{even} +  \epsilon (V_{odd} + V_{contact})$ |

```python
# epsilon is introduced for making cp-odd perturbative, it will be removed before calculating phase shifts.

def pot_total(s1,l1,j1,s2,l2,j2,i1,i2,index):
    
    if index == 0:
        return pot_even_dict(s1,l1,j1,s2,l2,j2,i1,i2)
    
    elif index == 1:
        return  epsilon * pot_cp_odd_dict(s1,l1,j1,s2,l2,j2,i1,i2)
    
    elif index == 2:
        return pot_even_dict(s1,l1,j1,s2,l2,j2,i1,i2) + epsilon * pot_cp_odd_dict(s1,l1,j1,s2,l2,j2,i1,i2) 
    
    elif index == 3:
        return pot_even_dict(s1,l1,j1,s2,l2,j2,i1,i2) + epsilon * pot_cp_odd_contact(s1,l1,j1,s2,l2,j2,i1,i2)
    
    elif index == 4:
        return pot_even_dict(s1,l1,j1,s2,l2,j2,i1,i2)  + epsilon * pot_cp_odd_dict(s1,l1,j1,s2,l2,j2,i1,i2) + epsilon * pot_cp_odd_contact(s1,l1,j1,s2,l2,j2,i1,i2) 
        
```

## Vtensor

### uncoupled phase

```python
# Vtensor without q0

def V_one_pion(s1,l1,j1,s2,l2,j2,index):
    
    Vtensor = {}

    for i1 in range(Np):        
        for i2 in range(Np):
            

            vtemp = pot_total(s1,l1,j1,s2,l2,j2,i1,i2,index)

            Vtensor[s1, l1, j1, s2, l2, j2, i1, i2] = vtemp


    return Vtensor

# Vtensor with q0 (so that when we change q0, we only need to do this code.)

def V_one_pion_q0(s1,l1,j1,s2,l2,j2,Vtensor,q0,index):

    pgrid[Np] = q0
    
    Vtensor2={}
    
    for i1 in range(Np):
        
        vtemp = pot_total(s1,l1,j1,s2,l2,j2,i1,Np,index)
        Vtensor2[s1,l1,j1,s2,l2,j2,i1,Np] = vtemp
        
        vtemp = pot_total(s1,l1,j1,s2,l2,j2,Np,i1,index)
        Vtensor2[s1,l1,j1,s2,l2,j2,Np,i1] = vtemp
        
    Vtensor2[s1,l1,j1,s2,l2,j2,Np,Np] = pot_total(s1,l1,j1,s2,l2,j2,Np,Np,index)
    
    for key in Vtensor:
        Vtensor2[key] = Vtensor[key]
    
    return Vtensor2
```

### coupled phase

```python
#Vtensor without q0

def Vtensor_calc(phase_list,index):

    len_phase = len(phase_list)

    Vtensor0 = {}

    for j_ in range(len_phase):
        for k_ in range(len_phase):

            [s1,l1,j1] = phase_list[j_]
            [s2,l2,j2] = phase_list[k_]

            Vtemp = V_one_pion(s1,l1,j1,s2,l2,j2,index)

            for key in Vtemp:
                Vtensor0[key] = Vtemp[key]

    return Vtensor0

# Vtensor with q0 (so that when we change q0, we only need to do this code.)

def Vtensor_q0_calc(phase_list,Vtensor0,q0,index):

    len_phase = len(phase_list)

    Vtensor = {}

    for j_ in range(len_phase):
        for k_ in range(len_phase):

            Vtemp = {}

            [s1,l1,j1] = phase_list[j_]
            [s2,l2,j2] = phase_list[k_]

            Vtemp = V_one_pion_q0(s1,l1,j1,s2,l2,j2,{},q0,index)

            for key in Vtemp:
                Vtensor[key] = Vtemp[key]
    
    for key in Vtensor0:
        Vtensor[key] = Vtensor0[key]

    return Vtensor
```

### to read $c_0$ data from CP-even and CP-odd

```python
# f1 = 1: read cp-even c0 data
# f1 = 2: read cp-odd c0 data

def c0_read(phase_list,f1=1,choice = False):
    
    
    if f1 == 1:
        folder = 'c0/c0_'
    elif f1 == 2:
        folder = 'c0/cpv_'
        
    # choose the eps_exp value while testing, 
    # set 'cpv_c0_choice' as a global variable to avoid 'choice' parameter
    
    global cpv_c0_choice
    if choice:
        cpv_c0_choice = int(input('choose eps value \n\n0: 0.01\n1: 0.11\n2: -0.09\n3: 0.00\n'))
    
    filename =''
    for phase in phase_list:
        filename += '{:}_'.format(phase_name(*phase))
    path = folder + filename + '.csv'
    
       
    if exists(path):
        
        pd1 = pd.read_csv(path,index_col=None)
        
        d1 = pd1.to_dict('list')
        
        d2 = {}
        
        if f1 ==1:
            
            for i in range(len(d1['lam'])):

                d2[d1['lam'][i]] = d1['c0'][i]
            return d2
        elif f1 ==2:
            
            col_name = ['e0_true = 0.01','e0_true = 0.11','e0_true = -0.09','not renorm']
            
            for i in range(len(d1['lam'])):
                
                lam = d1['lam'][i]
                col = col_name[cpv_c0_choice]
                e0_choice = d1[col][i]
                
                d2[lam] = e0_choice
            return d2
                
    # set c0 = 0
    else:
        
        d1 ={}
        for lam in np.arange(2,30.1,1):
            d1[lam] = 0
        return d1
        
```

where,
$V(p',p) \rightarrow e^{-p^4/\Lambda^4}(V_{index})e^{-p^{'4}/\Lambda^4}$

```python
def pot_lambda(Vtensor,lam,index):

    v_keys = Vtensor.keys()
    V_lambda_tensor = {}
    
    lam = round(lam,2)
    
    c0_1S0 = c0_read([[0,0,0]])
    c0_3P0 = c0_read([[1,1,0]])
    c0_3S1_3D1 = c0_read([[1,0,1],[1,2,1]])
    c0_3P2 = c0_read([[1,1,2]])
    c0_3D2 = c0_read([[1,2,2]])
    
    cpv_c0 = c0_read([[0,0,0],[1,1,0]],2)
    
    for elem in v_keys:
        
        [s1,l1,j1,s2,l2,j2,i1,i2] = elem
        
        V_even_contact = 0
        
        if index != 1:

            if [s1,l1,j1] == [0,0,0] and [s2,l2,j2] == [0,0,0]:     # 1S0-1S0

                c0 = c0_1S0[lam]
                V_even_contact =  pot_even_contact(*elem,c0)

            elif [s1,l1,j1] == [1,1,0] and [s2,l2,j2] == [1,1,0]:   # 3P0-3P0

                c0 = c0_3P0[lam]
                V_even_contact =  pot_even_contact(*elem,c0)

            elif [s1,l1,j1] == [1,0,1] and [s2,l2,j2] == [1,0,1]:   # 3S1-3S1

                c0 = c0_3S1_3D1[lam]
                V_even_contact =  pot_even_contact(*elem,c0)

            elif [s1,l1,j1] == [1,1,2] and [s2,l2,j2] == [1,1,2]:   # 3P2-3P2

                c0 = c0_3P2[lam]
                V_even_contact =  pot_even_contact(*elem,c0)

            elif [s1,l1,j1] == [1,2,2] and [s2,l2,j2] == [1,2,2]:   # 3D2-3D2

                c0 = c0_1D2[lam]
                V_even_contact =  pot_even_contact(*elem,c0)
        
        V_odd_contact = 0
        
        if index==2 and exists('c0/cpv_1S0_3P0_.csv'):
            
            V_odd_contact = cpv_c0[lam] * pot_cp_odd_contact(s1,l1,j1,s2,l2,j2,i1,i2) * epsilon

        
        
        V_lambda_tensor[elem] = (
                                    exp( -(pgrid[i1]/ (lam * hc) )**4 )
                                    *   (
                                            Vtensor[elem] + V_even_contact + V_odd_contact
                                        )
                                    * exp( -(pgrid[i2]/ (lam * hc) )**4 )
                                )
                                
    
    return V_lambda_tensor
```
